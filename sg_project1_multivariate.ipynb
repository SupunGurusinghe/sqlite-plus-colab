```python
c = conn.cursor()

def null_count(table_n):

  table_name = table_n

  c.execute("SELECT name FROM pragma_table_info(?) ORDER BY cid", [table_name])
  names = c.fetchall()

  result_list = []
  for name in names:
    result_list.append(name[0])

  c.execute(f'SELECT * FROM {table_name}')
  results = c.fetchall()

  df = pd.DataFrame(results, columns= result_list)
  df.head()

  null_count_matrix = []
  i = 0
  for col in df.columns:
    null_count_matrix.append([])

    null_count_matrix[i].append(col)
    null_count_matrix[i].append(df[col].isna().sum())
    null_count_matrix[i].append(df[df[col].isnull()].index.tolist())
    i += 1

  null_df = pd.DataFrame(null_count_matrix, columns= ['Column', 'NullCount', 'NullIdList'])

  return null_df

c.close()
```

```python
c = conn.cursor()

null_df = null_count('storefront1')

c.close()
```

```python
c = conn.cursor()

import plotly.express as px

def multivariate(table1, table2, df, df1):
  # get to a dataframe
  table1_output = null_df = null_count(table1)
  table2_output = null_df = null_count(table2)


  print("-"*100)
  print("Dataframe difference -- \n")
  print(df.compare(df1))

  # Keep equal values
  #print("-"*5)
  #print("Dataframe difference keeping equal values -- \n")
  #print(df.compare(df1, keep_equal=True))

  print("-"*100)
  print("Dataframe difference keeping same shape -- \n")
  print(df.compare(df1, keep_shape=True))
  
  # Keep equal values
  #print("-"*5)
  #print("Dataframe difference keeping same shape and equal values -- \n")
  #print(df.compare(df1, keep_shape=True, keep_equal=True))


  # compare null by ploting
  fig1 = px.bar(table1_output, 
                x='Column', 
                y='NullCount', 
                height=400, 
                title='Before')
  fig1.show()

  fig2 = px.bar(table2_output, 
                x='Column', 
                y='NullCount', 
                height=400, 
                title='After')
  fig2.show()

c.close()
```

```python
c = conn.cursor()

multivariate('storefront1', 'storefront2', df1, df2)

c.close()
```
